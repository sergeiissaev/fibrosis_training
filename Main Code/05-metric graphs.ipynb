```python
bs = 1
n_epochs = 150
train_size = 1753
bat_per_epo = train_size / bs
n_steps = bat_per_epo * n_epochs
print('There are', int(bat_per_epo), 'batches per epoch.\nThe program ran for', int(n_steps), 'steps.')
run_name = 'test0658'
```

# End of Edit Here

### Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import pickle
```

### Load Data

```python
with open("/home/sergei/projects/def-rogertam/sergei/micro_scans/models/" + run_name + "/dlr.txt", "rb") as fp:
    Disc_real = pickle.load(fp)

with open("/home/sergei/projects/def-rogertam/sergei/micro_scans/models/" + run_name + "/dlf.txt", "rb") as fp:
    Disc_fake = pickle.load(fp)

with open("/home/sergei/projects/def-rogertam/sergei/micro_scans/models/" + run_name + "/gl.txt", "rb") as fp:
    Gen = pickle.load(fp)

Disc_loss_real = []
Disc_loss_fake = []
Gen_loss = []
cut = bat_per_epo//bs


for i in range(len(Disc_real)):
    if  i % cut == 0:
        Gen_loss.append(Gen[i])
        Disc_loss_fake.append(Disc_fake[i])
        Disc_loss_real.append(Disc_real[i])
print(f"Gen loss has {len(Gen_loss)} items.")
print(len(Gen_loss) == n_epochs)
```

### Plot Graphs

```python
newfolder = '/home/sergei/projects/def-rogertam/sergei/micro_scans/models/' + run_name + '/report/'
!mkdir $newfolder
sty = 'seaborn'
epoch = range(0, len(Gen_loss))
mpl.style.use(sty)
fig, ax = plt.subplots(figsize=(6, 4))
plt.plot(epoch, Disc_loss_real, 0.3, linewidth=0.8, color='#9b0000', marker='.')
plt.title('Discriminator loss real')
plt.xlabel('epoch')
plt.ylabel('Loss')
plt.savefig('/home/sergei/projects/def-rogertam/sergei/micro_scans/models/' + run_name + '/report/discriminator_real_loss.jpg')
plt.show()
fig, ax = plt.subplots(figsize=(6, 4))
plt.plot(epoch, Disc_loss_fake, 0.3, linewidth=0.8, color='#1562ff' ,marker='.')
plt.title('Discriminator loss fake')
plt.xlabel('epoch')
plt.ylabel('Loss')
plt.savefig('/home/sergei/projects/def-rogertam/sergei/micro_scans/models/' + run_name + '/report/discriminator_fake_loss.jpg')
plt.show()
fig, ax = plt.subplots(figsize=(6, 4))
plt.plot(epoch, Gen_loss, 0.6,linewidth=0.8, color='#8a3ac6' ,marker='.')
plt.title('Generator loss')
plt.xlabel('epoch')
plt.ylabel('loss')
plt.savefig('/home/sergei/projects/def-rogertam/sergei/micro_scans/models/' + run_name + '/report/gen_loss.jpg')
plt.show()
```
